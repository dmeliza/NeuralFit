```python
def euler ( f, dt, tfinal, x0):
    N = int(tfinal / dt)
    X = np.zeros((N, x0.size))
    state = x0.copy()
    X[0,:] = state
    for i in range (0, N):
        state = state + (f(state, i * dt) * dt)
        X[i,:] = state
    return X
```

```python
def f(state, t):
    x, y, z, rho, sigma, beta = state  # unpack the augmented state vector
    return np.array((sigma * (y - x), x * (rho - z) - y, x * y - beta * z, 0, 0, 0))  # derivatives
```

```python
 def normpdf(mu, x, sd):
    var = float(sd)**2
    pi = np.pi
    denom = (2*pi*var)**.5
    return (math.exp(-(float(x)-float(mu))**2/(2*var)))/denom
```

```python
# CLACULATES EFFECTIVE SAMPLE SIZE

def ess(weights):   
    s1 = np.exp(weights).sum()
    s2 = np.exp(2 * weights).sum()
    
    ess_D = np.exp(2 * np.log(s1) - np.log(s2))
    return ess_D
```

```python
def stepIntegrate(pn, p_states, iterations, step, p_weights, data, noise, tolerance, sampling_threshold, resample_count):     
    
    loglikelihoods = np.zeros(pn)
    
    for p in range (0, pn):
        
            p_states[p] = p_states[p] + (f(p_states[p], iterations * dt) * dt) + noise[p] 
            
            o_val = [p_states[p][0], p_states[p][1], p_states[p][2]]
           
            loglikelihood =  (np.log(normpdf(data[0][step / dt], o_val[0], tolerance)) +
                           np.log(normpdf(data[1][step / dt], o_val[1], tolerance)) +
                           np.log(normpdf(data[2][step / dt], o_val[2], tolerance)))
    
            loglikelihoods[p] = loglikelihood
              
    ## UPDATE, PROCESS WEIGHTS
    
    p_weights = p_weights + loglikelihoods
 
    ## NORMALISING WEIGHTS FOR RESAMPLING 

    max_weight = max(p_weights) 
    p_weights -= max_weight 
        
    ## RESAMPLING 
                                  
    curr_ess = ess(p_weights)
    
    if (curr_ess < (sampling_threshold * pn)):
    
        resample_count = resample_count + 1
        
        p_weights = np.exp(p_weights)
         
        p_weights /= np.sum(p_weights)
        
        p_states.append(0)
        w = p_weights.tolist()
        w.append(0.0)
        
        p_states = (np.random.choice(p_states, pn, p=w))
        p_states = p_states.tolist()
        
        p_weights[:] = 0
     
    #print "Particle states:", p_states, "\n"
    #print "Particle weights:", p_weights, "\n"
    
    return p_states, p_weights, resample_count
```

```python
def pf(particles, time, iterations, dt, start_pos, data, s_noise, d_noise, tolerance):
    
    ## INITIALIZING PARTICLE FILTER
    
    pn = particles
    p_states = []
    traces = []
    step = 0.0
    sampling_threshold = float(0.04)
     
    # ASSIGNING RANDOM STARTING POSITIONS FOR EACH PARAMETER, FOR ALL PARTICLES
    
    x_starts = np.random.normal(start_pos[0], s_noise, pn) 
    y_starts = np.random.normal(start_pos[1], s_noise, pn) 
    z_starts = np.random.normal(start_pos[2], s_noise, pn) 
    rho_starts = np.random.normal(start_pos[3], s_noise, pn) 
    sigma_starts = np.random.normal(start_pos[4], s_noise, pn)
    beta_starts = np.random.normal(start_pos[5], s_noise, pn)
    
    for h in range (0, pn):     
        state0 = [x_starts[h], y_starts[h], z_starts[h], rho_starts[h], sigma_starts[h], beta_starts[h]]
        p_states.append(state0)  

    ## INTEGRATING LORENZ SYSTEM STEP BY STEP
    
    p_weights = np.zeros(shape=(pn))
    resample_count = 0
    
    for i in range (0, iterations):
        
        #print "\ncurrently step is :", step, "------------------------------------------"
        
        ## CALCULATING SYSTEM NOISE
        
        noise = np.random.normal(0, d_noise, (pn, 6))
        noise = np.round(noise, decimals=4)
        
        ## CALLING STEP INTEGRATOR
        
        pstates , weights, r_count  = stepIntegrate(pn, p_states, iterations, step, p_weights, data, noise, tolerance,sampling_threshold, resample_count)
        
        p_weights = weights
        p_states = pstates
        resample_count = r_count
        
        traces.append(np.array(p_states))         
        
        step = step + dt
        
    return np.array(p_states), np.array(weights), np.array(traces), resample_count
```
